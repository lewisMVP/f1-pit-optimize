```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

tire_wear = np.linspace(1, 0, 100)
lap_time = 85 + 15 * (1 - tire_wear) + np.random.normal(0, 1, 100)

X = tire_wear.reshape(-1, 1)
y = lap_time

model = LinearRegression()
model.fit(X, y)

predicted_lap_time = model.predict(X)

plt.scatter(tire_wear, lap_time, color='blue', label='Actual data')
plt.plot(tire_wear, predicted_lap_time, color='red', label='Linear regression line')
plt.xlabel('Tire Wear (1 = new, 0 = worn)')
plt.ylabel('Lap Time (seconds)')
plt.title('Relationship between Tire Wear and Lap Time')
plt.legend()
plt.grid(True)
plt.show()

print(f"Model coefficient (slope): {model.coef_[0]}")
print(f"Model intercept: {model.intercept_}")
```
